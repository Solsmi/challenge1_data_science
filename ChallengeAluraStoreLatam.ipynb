Durante este desafío, ayudarás al Sr. Juan a decidir qué tienda de su cadena Alura Store debe vender para iniciar un nuevo emprendimiento. ##Extracción de datos

### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

##Análisis de datos

###1. Facturación



En este primer análisis, debes calcular el ingreso total de cada tienda. Sumando los valores de la columna Precio de cada conjunto de datos de la tienda para estimar los ingresos.

```python
suma_total_precio = tienda["Precio"].sum()
suma_total_precio2 = tienda2["Precio"].sum()
suma_total_precio3 = tienda3["Precio"].sum()
suma_total_precio4 = tienda4["Precio"].sum()
```

```python
print(f'La facturación total de la tienda 1 es: ${suma_total_precio}')
print(f'La facturación total de la tienda 2 es: ${suma_total_precio2}')
print(f'La facturación total de la tienda 3 es: ${suma_total_precio3}')
print(f'La facturación total de la tienda 4 es: ${suma_total_precio4}')
```

### 2. Ventas por categoría

En este debes calcular la cantidad de productos vendidos por categoría en cada tienda. La idea es agrupar los datos por categoría y contar el número de ventas de cada tipo, mostrando las categorías más populares de cada tienda.

```python
print("Ventas y Facturación por Categoría Tienda 1:")
ventas_y_facturacion_tienda1 = tienda.groupby('Categoría del Producto').agg(Total_Facturacion=('Precio', 'sum'), Cantidad_Ventas=('Producto', 'count')).sort_values(by='Cantidad_Ventas', ascending=False)
display(ventas_y_facturacion_tienda1)
print("\n")

print("Ventas y Facturación por Categoría Tienda 2:")
ventas_y_facturacion_tienda2 = tienda2.groupby('Categoría del Producto').agg(Total_Facturacion=('Precio', 'sum'), Cantidad_Ventas=('Producto', 'count')).sort_values(by='Cantidad_Ventas', ascending=False)
display(ventas_y_facturacion_tienda2)
print("\n")

print("Ventas y Facturación por Categoría Tienda 3:")
ventas_y_facturacion_tienda3 = tienda3.groupby('Categoría del Producto').agg(Total_Facturacion=('Precio', 'sum'), Cantidad_Ventas=('Producto', 'count')).sort_values(by='Cantidad_Ventas', ascending=False)
display(ventas_y_facturacion_tienda3)
print("\n")

print("Ventas y Facturación por Categoría Tienda 4:")
ventas_y_facturacion_tienda4 = tienda4.groupby('Categoría del Producto').agg(Total_Facturacion=('Precio', 'sum'), Cantidad_Ventas=('Producto', 'count')).sort_values(by='Cantidad_Ventas', ascending=False)
display(ventas_y_facturacion_tienda4)
```

### 3. Calificación promedio de la tienda


En este paso, debes calcular las calificaciones promedio de los clientes para cada tienda. El objetivo es conocer la satisfacción del cliente con los productos vendidos.

```python
calificacion_promedio_tienda1 = tienda['Calificación'].mean()
calificacion_promedio_tienda2 = tienda2['Calificación'].mean()
calificacion_promedio_tienda3 = tienda3['Calificación'].mean()
calificacion_promedio_tienda4 = tienda4['Calificación'].mean()
print(f'La calificación promedio de la tienda 1 es: {calificacion_promedio_tienda1:.2f}')
print(f'La calificación promedio de la tienda 2 es: {calificacion_promedio_tienda2:.2f}')
print(f'La calificación promedio de la tienda 3 es: {calificacion_promedio_tienda3:.2f}')
print(f'La calificación promedio de la tienda 4 es: {calificacion_promedio_tienda4:.2f}')
```

### 4. Productos más y menos vendidos

En este paso, debes identificar qué productos fueron los más vendidos y los menos vendidos en cada tienda. Visualiza los resultados para que quede claro qué productos destacaron en ventas en cada tienda.

```python
def analizar_ventas_productos(df, nombre_tienda):
    ventas_productos = df['Producto'].value_counts()

    print(f"\n--- {nombre_tienda} ---")

    if not ventas_productos.empty:
        # Productos más vendidos
        print("Productos más vendidos:")
        top_n = min(3, len(ventas_productos))
        productos_mas_vendidos = ventas_productos.head(top_n).sort_values(ascending=False)
        for producto, cantidad in productos_mas_vendidos.items():
            print(f"  - {producto}: {cantidad} unidades")

        # Productos menos vendidos
        if len(ventas_productos) > 0:
            print("\nProductos menos vendidos:")
            bottom_n = min(3, len(ventas_productos))
            productos_menos_vendidos = ventas_productos.tail(bottom_n).sort_values(ascending=True)
            for producto, cantidad in productos_menos_vendidos.items():
                print(f"  - {producto}: {cantidad} unidades")
        else:
            print("No hay productos menos vendidos para analizar.")

    else:
        print("No hay datos de productos para analizar.")

analizar_ventas_productos(tienda, "Tienda 1")
analizar_ventas_productos(tienda2, "Tienda 2")
analizar_ventas_productos(tienda3, "Tienda 3")
analizar_ventas_productos(tienda4, "Tienda 4")
```

### 5. Envío promedio por tienda

En este paso, debes calcular el costo de envío promedio para cada tienda. El objetivo es comprender cuánto se gasta, en promedio, en el envío de cada tienda.  Sabiendo que el cliente paga el envío.

```python
costo_envio_promedio_tienda1 = tienda['Costo de envío'].mean()
costo_envio_promedio_tienda2 = tienda2['Costo de envío'].mean()
costo_envio_promedio_tienda3 = tienda3['Costo de envío'].mean()
costo_envio_promedio_tienda4 = tienda4['Costo de envío'].mean()
print(f'El costo de envío promedio de la tienda 1 es: ${costo_envio_promedio_tienda1:.2f}')
print(f'El costo de envío promedio de la tienda 2 es: ${costo_envio_promedio_tienda2:.2f}')
print(f'El costo de envío promedio de la tienda 3 es: ${costo_envio_promedio_tienda3:.2f}')
print(f'El costo de envío promedio de la tienda 4 es: ${costo_envio_promedio_tienda4:.2f}')
```

#### **Productos más vendidos y menos vendidos**

```python
import pandas as pd

# 1. Inicializar una lista vacía llamada products_for_chart.
products_for_chart = []

# 2. Definir n_products como 3.
n_products = 3

# 3. Crear una lista de los DataFrames de las tiendas y sus nombres.
store_dfs = [tienda, tienda2, tienda3, tienda4]
nombres_tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

# 4. Recorrer cada DataFrame de tienda y su nombre correspondiente:
for i, df in enumerate(store_dfs):
    store_name = nombres_tiendas[i]

    # a. Calcular el conteo de ventas para cada producto en la tienda actual.
    ventas_productos = df['Producto'].value_counts()

    # b. Seleccionar los n_products productos más vendidos.
    top_products = ventas_productos.head(n_products).reset_index()
    top_products.columns = ['Producto', 'Cantidad_Ventas']
    top_products['Tienda'] = store_name
    top_products['Tipo'] = 'Más Vendido'
    products_for_chart.append(top_products)

    # c. Seleccionar los n_products productos menos vendidos.
    least_products = ventas_productos.tail(n_products).reset_index()
    least_products.columns = ['Producto', 'Cantidad_Ventas']
    least_products['Tienda'] = store_name
    least_products['Tipo'] = 'Menos Vendido'
    products_for_chart.append(least_products)

# 5. Concatenar todos los DataFrames en products_for_chart en un único DataFrame unificado.
products_for_chart_df = pd.concat(products_for_chart, ignore_index=True)

print("DataFrame unificado 'products_for_chart_df' creado exitosamente.")
print(products_for_chart_df.head())
```

Ubicacion de ventas por tienda

```python
def get_top_bottom_locations(df, store_name, n=3):
    # Ordenar por 'Precio' (ventas totales) para encontrar los de mejor y peor rendimiento
    df_sorted = df.sort_values(by='Precio', ascending=False)

    top_n_locations = df_sorted.head(n).copy()
    top_n_locations['Tienda'] = store_name
    top_n_locations['Tipo'] = 'Mayor Venta'

    bottom_n_locations = df_sorted.tail(n).copy()
    bottom_n_locations['Tienda'] = store_name
    bottom_n_locations['Tipo'] = 'Menor Venta'

    return top_n_locations, bottom_n_locations

# Calcular las ventas por ubicación para cada tienda
ventas_por_lugar_tienda1 = tienda.groupby('Lugar de Compra').agg(Precio=('Precio', 'sum')).reset_index()
ventas_por_lugar_tienda2 = tienda2.groupby('Lugar de Compra').agg(Precio=('Precio', 'sum')).reset_index()
ventas_por_lugar_tienda3 = tienda3.groupby('Lugar de Compra').agg(Precio=('Precio', 'sum')).reset_index()
ventas_por_lugar_tienda4 = tienda4.groupby('Lugar de Compra').agg(Precio=('Precio', 'sum')).reset_index()

# Aplicar la función a los datos de ventas de cada tienda
top_tienda1, bottom_tienda1 = get_top_bottom_locations(ventas_por_lugar_tienda1, 'Tienda 1')
top_tienda2, bottom_tienda2 = get_top_bottom_locations(ventas_por_lugar_tienda2, 'Tienda 2')
top_tienda3, bottom_tienda3 = get_top_bottom_locations(ventas_por_lugar_tienda3, 'Tienda 3')
top_tienda4, bottom_tienda4 = get_top_bottom_locations(ventas_por_lugar_tienda4, 'Tienda 4')

# Consolidar todos los de mejor y peor rendimiento en una sola lista de DataFrames
ventas_por_lugar_consolidado = [
    top_tienda1, bottom_tienda1,
    top_tienda2, bottom_tienda2,
    top_tienda3, bottom_tienda3,
    top_tienda4, bottom_tienda4
]

# Concatenar todo en un único DataFrame para facilitar el análisis y la visualización
df_ventas_por_lugar = pd.concat(ventas_por_lugar_consolidado, ignore_index=True)

print("DataFrame consolidado de ubicaciones de mejor y peor rendimiento:")
print(df_ventas_por_lugar.head())
```

Neiva, Inírida, Valledupar), la cadena debería definir una estrategia clara: ¿Se invierte para crecer en estas zonas, se aceptan como mercados secundarios de bajo volumen, o se considera una reestructuración de la presencia en ellas? Estas recomendaciones buscan capitalizar las fortalezas actuales de Alura Store y abordar las áreas de mejora identificadas a través de un enfoque basado en datos y adaptado a las particularidades de cada tienda.
